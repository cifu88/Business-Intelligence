# Covariance Background:
You are given data on the SAT reading and writing scores of several students from our lesson on cross tables and scatter plots:

```python
write = [344,383,611,713,536]
read = [378,349,503,719,503]
```

## Task 1:
Determine if this is sample or population.

## R:
This is a sample because in the background it says "several students".
- The covariance formula is : SUM((x - x.mean)*(y - y.mean))/n-1

## Task 2:
Calculate the covariance of the two datasets:

```python
import statistics as st

cov = st.covariance(write, read)
print(cov)
```

## Task 3:
Plot the data on scatter plot and using your previous knowledge comment on whether there is a noticeable relationship between the two variables.

```python
import seaborn as sbn

sbn.scatterplot(x=write, y=read)
print("Whith the scatter plot is possible to see that there is a relationship between this two variables because when the writting score grows up the reading score do the same.")
```
